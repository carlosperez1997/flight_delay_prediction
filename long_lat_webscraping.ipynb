```python
airports_df = pd.read_csv(os.path.join('/Users/carlosperezricardo/Desktop/TFM/airlines_airports', 'L_AIRPORT_ID.csv'))
```

```python
airports_df = airports_df[['Code','Description','City_State','City','State_Country','lat','lon']]
```

```python
airports_df['City_State'] = [ x.split(':')[0].strip() for x in airports_df['Description']]
airports_df['City'] =  [ x.split(',')[0].strip() for x in airports_df['City_State']]
airports_df['State_Country'] =  [ x.split(',')[1].strip() if ',' in x else '' for x in airports_df['City_State'] ]
```

```python
if 'lat' not in airports_df.columns:
    airports_df['lat'] = 0
    airports_df['lon'] = 0
```

```python
for idx, airport in airports_df.iterrows():
    city_state = airport['City_State']
    state = airport['State_Country']
    lat = airport['lat']
    lon = airport['lon']

    if len(str(state)) == 2 and lon == 0 and lat == 0:
        driver.find_element_by_name("address").send_keys(city_state)
        driver.find_element_by_name("load_input_btn").click()

        time.sleep(12)

        content = driver.page_source
        soup = BeautifulSoup(content)

        lat_long = soup.findAll(class_="value green")

        lat = lat_long[0].text.split('°')[0]
        lon = lat_long[1].text.split('°')[0]

        airports_df['lat'].iloc[idx] = lat
        airports_df['lon'].iloc[idx] = lon

        text_area = driver.find_element_by_class_name('address')

        for i in range(len(city_state)):
            text_area.send_keys(Keys.BACKSPACE)
    
```
