```python
class FeatureGenerator_adapted(object):
    
    '''
    This is a helper class that takes a df and a list of features and creates sum, mean, 
    lag features and variation (change over month) features.
    
    '''
    
    def __init__(self, full_df,  gb_list):
        
        '''
        Constructor of the class.
        gb_list is a list of columns that must be in full_df.
        '''
        
        self.full_df = full_df
        self.gb_list = gb_list
        # joins the gb_list, this way we can dinamically create new columns
        # ["date, "shop_id] --> date_shop_id
        self.objective_column_name = "_".join(gb_list)
            
    def generate_gb_df(self, calculations):
        '''
        This function thakes the full_df and creates a groupby df based on the gb_list.
        It creates 2 columns: 
            1. A sum column for every date and gb_list
            2. Mean columns for every_date and gb_list
            
        The resulting df (gb_df_) is assigned back to the FeatureGenerator class as an attribute.
        '''

        def my_agg(full_df_, args):
            
            '''
            This function is used to perform multiple operations over a groupby df and returns a df
            without multiindex.
            '''
            
            names = {}
            for key, value in calculations.items():
                if value[1] == 'sum':
                    names['{}_{}'.format(args,key)] = full_df_[value[0]].sum() 
                elif value[1] == 'mean':
                    names['{}_{}'.format(args,key)] = full_df_[value[0]].mean() 
                elif value[1] == 'min':
                    names['{}_{}'.format(args,key)] = full_df_[value[0]].min() 
                elif value[1] == 'max':
                    names['{}_{}'.format(args,key)] = full_df_[value[0]].max() 
                elif value[1] == 'median':
                    names['{}_{}'.format(args,key)] = full_df_[value[0]].median() 
                
            return pd.Series(names, index = [key for key in names.keys()])
        
        # the args is used to pass additional argument to the apply function
        gb_df_ = self.full_df.groupby(self.gb_list).apply(my_agg, args = (self.objective_column_name)).reset_index()

        self.gb_df_ = gb_df_
        

    def generate_shift_features(self, suffixes, grados_p, grados_q):
            
        '''
        This function is a helper function that takes the gb_df_ and a suffix (sum or mean) and creates the
        additional features.
        '''

        # dinamically creates the features
        for suffix in suffixes:
            name_ = self.objective_column_name + "_" + suffix
            
            for grado in grados_p:
                self.gb_df_['{}_shift_{}'.format(name_, grado)] =\
                self.gb_df_.groupby(self.gb_list[1:])[name_].transform(lambda x: x.shift(grado))

            for grado in grados_q:
                self.gb_df_['{}_var_pct_{}'.format(name_, grado)] =\
                self.gb_df_.groupby(self.gb_list[1:])[name_].transform(lambda x: (x.shift(1) - x.shift(grado+1))/x.shift(grado+1))
            
        self.gb_df_.fillna(-1, inplace = True)

        self.gb_df_.replace([np.inf, -np.inf], -1, inplace = True)
            
        return self.gb_df_
```

```python
calculations = {'sum':['ARR_DELAY_15','sum'],
                'mean':['ARR_DELAY_15','mean']
               }

# Grados p y q sacados del EDA 
grados_pp = [1,2,3] 
grados_qq = [1,2]

#st = time.time()

gb_list = ["FL_DATE_month", "OP_UNIQUE_CARRIER"]

fe_generator = FeatureGenerator_adapted(full_df = full_df, gb_list = gb_list)

fe_generator.generate_gb_df(calculations)

carrier_month_features = fe_generator.generate_shift_features( suffixes = calculations.keys(), grados_p = grados_pp, grados_q = grados_qq)

#et = time.time()

#(et - st)/60
```
